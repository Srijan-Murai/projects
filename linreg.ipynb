```python
import matplotlib.pyplot as plt
from random import uniform, randint, random
from time import sleep
from IPython.display import clear_output as clear
import numpy as np
```

```python
# equation to generate points in a recognizable pattern
e = (randint(-2,2), randint(-50,50) )

n = 30 # number of points CHANGEBLE

x = []
y = []
for i in range(n):
    a = randint(-10,100)
    b = e[0]*a+e[1]+ randint(-20,20)
    x.append(a)
    y.append(b)

plt.figure(figsize=(6, 6))
plt.xlim(0, 100)
plt.ylim(-100, 100)
plt.scatter(x, y, color='blue', alpha=0.5)


xy = [x[i]*y[i] for i in range(n)]
x2 = [x[i]**2 for i in range(n)]

def eq(sx,sy,sxy,sx2):
    m = round((n*sxy-sx*sy)/(n*sx2-sx**2),5)
    b = round((sy-m*sx)/n,5)
    return m,b
# x,y
```

```python
l = 1
while l <= n:
    plt.figure(figsize=(6, 6))
    plt.xlim(0, 100)
    plt.ylim(-100, 100)

    tx = x[:l]
    ty = y[:l]
    txy = xy[:l]
    tx2 = x2[:l]
    
    sx = sum(tx[:l])
    sy = sum(ty[:l])
    sxy = sum(txy[:l])
    sx2 = sum(tx2[:l])

    slope, intercept = eq(sx,sy,sxy,sx2)

    currx = np.linspace(0, 100, 100)
    curry = slope*currx + intercept

    plt.scatter(x, y, color='blue', alpha=0.5)
    plt.plot(currx, curry, color="blue")
    plt.title(f"y = {slope}x + {intercept}")
    plt.show()
    sleep(0.5)
    clear(wait=True)

    l+=1
```
